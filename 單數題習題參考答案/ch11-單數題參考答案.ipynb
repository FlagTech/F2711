```python
import numpy as np
import matplotlib.pyplot as plt
```

### 11.1 Matplotlib 繪圖的基本認識

#### 第1題

```python
#(a)
x=np.linspace(-7,4)
y=x**4+6*x**3+7*x+3
fig=plt.figure(figsize=(5,4))
ax=fig.add_subplot()
ax.plot(x,y,'r')
plt.show()
```

```python
#(b)
x=np.linspace(-np.pi,2*np.pi,256)
y=6*np.sin(x+3)*np.cos(x)
fig=plt.figure(figsize=(5,4))
ax=fig.add_subplot()
ax.plot(x,y,'r')
plt.show()
```

```python
#(b)
x=np.linspace(-3,6,128)
y=(x+3)/(x**2+1)
fig=plt.figure(figsize=(5,4))
ax=fig.add_subplot()
ax.plot(x,y,'r')
plt.show()
```

#### 第3題

```python
x=np.linspace(0,4*np.pi,128)
f=np.sin(x)/(x+1)

fig=plt.figure(figsize=(5,4))
ax=fig.add_subplot()
ax.plot(x,f,'r')
plt.savefig('my_plot.png',dpi=72,transparent=False)
plt.show()
```

### 11.2 二維圖形的修飾

#### 第5題

```python
x=np.linspace(0,2*np.pi,128)
y1=np.sin(x)
y2=np.sin(2*x)
y3=np.sin(3*x)
y4=np.sin(4*x)

fig,ax=plt.subplots(2,2,figsize=(6,4))
ax[0,0].plot(x,y1,'r')
ax[0,1].plot(x,y2,'g')
ax[1,0].plot(x,y3,'b')
ax[1,1].plot(x,y4,'m')
plt.show()
```

#### 第7題

```python
x=np.linspace(1,1000,160)
y=x*np.log10(3*x)*np.sin(np.sqrt(x))
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xscale('log')
ax.grid()
plt.show()
```

#### 第9題

```python
x=np.linspace(1,100,132)
y=x*np.exp(x)/(x**2+1)

fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid()
plt.show()
```

### 11.3 填滿繪圖與極坐標繪圖

#### 第11題

```python
x=np.linspace(0,np.pi,32)
y=np.exp(-0.5*x)*np.cos(3*x)

fig,ax=plt.subplots()
ax.plot(x,y,'ro--',markerfacecolor='yellow',markersize=8)
ax.set_title('My plot',fontsize=18)
ax.grid()
plt.show()
```

#### 第13題

```python
x=np.linspace(0,2*np.pi,120)
r1=np.sin(3*x)
r2=np.cos(np.sin(6*x))
fig,ax=plt.subplots(subplot_kw={'projection':'polar'})  # keyword pass to subplot() in dict from
ax.plot(x,r1)
ax.plot(x,r2)
ax.set_rticks([-1,0,1])
plt.show()
```

### 11.4 統計繪圖

#### 第15題

```python
data=np.random.normal(1,5,3600)
fig,ax=plt.subplots()
ax.hist(data,bins=20,rwidth=0.8)
plt.show()
```

#### 第17題

### 11.5 等高線圖與三維繪圖

#### 第19題

```python
x=np.random.rand(1000)*6
y=np.random.rand(1000)*6
z=np.random.normal(5,8,1000)
fig,ax=plt.subplots(subplot_kw={'projection':'3d'})
ax.scatter(x,y,z,c=z,cmap='jet')                
```

#### 第21題

```python
x=np.linspace(-2,2,128)
y=np.linspace(-2,2,128)
XX,YY=np.meshgrid(x,y)
ZZ=np.log10(1+XX**2+YY**2)

fig,ax= plt.subplots(1,2,figsize=(10,4),subplot_kw={'projection':'3d'})
ax[0].contour3D(XX,YY,ZZ,[0.1,0.3,0.5],linewidths=2,colors='black')
ax[1].plot_surface(XX,YY,ZZ,cmap='jet',alpha=0.7)

plt.show()
```

### 11.6 動畫的製作

#### 第23題

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
x=np.linspace(0,2*np.pi,256)
r=np.sin(0)
fig,ax=plt.subplots(subplot_kw={'projection':'polar'})  # keyword pass to subplot() in dict from
line,= ax.plot([],[])
dot,=ax.plot([],[],'ro', markersize=12)
ax.set_rticks([-1,-0.5,0,0.5,1])

plt.close()
def animate(i):
    line.set_data(x[:i],np.sin(6*x[:i]))
    dot.set_data(x[i],np.sin(6*x[i]))

ani=FuncAnimation(fig=fig,
                  func=animate,
                  frames=256,
                  interval=60, blit=False)
HTML(ani.to_html5_video())
```
